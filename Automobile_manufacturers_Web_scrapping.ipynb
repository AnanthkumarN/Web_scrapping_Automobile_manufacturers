```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_automotive_manufacturers_by_production'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[5]
```

```python
table
```

```python
Automobile_titles = table.find_all('th')
```

```python
Automobile_titles
```

```python
Automobile_table_titles = [title.text.strip() for title in Automobile_titles]
```

```python
Automobile_table_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = Automobile_table_titles)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  length = len(df)
  df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv('Automobile_manufacturers.csv')
```
